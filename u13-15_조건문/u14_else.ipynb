### 3항 연산자

```python
x = 1
if x % 2 == 0:
  y = 'even'
else:
  y = 'odd'
y
```

```python
y = 'even' if x % 2 == 0 else 'odd'
y
```

```python
score = int(input('점수입력 > '))

if score >= 90:
  print('A')
elif score >= 80:
  print('B')
elif score >= 70:
  print('C')
elif score >= 60:
  print('D')
else:
  print('F')


if score < 60:
  print('F')
else:
  if score < 70:
    print('D')
  else:
    if score < 80:
      print('C')
    else:
      if score < 90:
        print('B')
      else:
        print('A')
```

```python
grade = 'A' if score >= 90 else \
          'B' if score >= 80 else \
            'C' if score >= 70 else \
              'D' if score >= 60 else 'F'
grade
```
